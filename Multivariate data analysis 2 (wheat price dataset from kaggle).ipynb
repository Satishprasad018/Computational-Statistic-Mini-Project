```python
import statistics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Daily_wheat_price.csv')
data.head(10)
```

```python
print(data.shape)
data.describe()
```

```python
mean_open = np.mean(data.open)
print("\nMean_open: ", mean_open)
mean_high = np.mean(data.high)
print("\nMean_high: ", mean_high)
mean_low = np.mean(data.low)
print("\nMean_low: ", mean_low)
mean_close = np.mean(data.close)
print("\nMean_close: ", mean_close)
```

```python
data.corr()
```

```python
print("% s"%(statistics.variance(data.open)))
print(" % s"%(statistics.variance(data.high)))
print("% s"%(statistics.variance(data.low)))
print("% s"%(statistics.variance(data.close)))
```

```python
cov_mat = np.stack((data.open,data.high, data.low,data.close), axis = 0)
print(np.cov(cov_mat))
```
